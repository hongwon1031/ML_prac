## 5. 그룹화
`groupby`의 결과는 **dictionary** 형태

```python
cust = pd.read_csv('sc_cust_info_txn_v1.5.csv')
cust
```

### 1. 단일 기준 그룹

```python
gender_group = cust.groupby('sex_type') # 딕셔너리 형태
gender_group    
```

```python
gender_group.count()
```

```python
gender_group.mean(numeric_only=True) #그룹별 평균 출력(수 형태만)
```

```python
gender_group.mean(numeric_only=True)[['r3m_avg_bill_amt']] # 특정 col의 평균만 볼때때
```

```python
gender_group.max(numeric_only=True)[['r6m_A_avg_arpu_amt','r6m_B_avg_arpu_amt']] # 특정 col의 최대값
```

### 2. 복수 기준 그룹

```python
multi_group = cust.groupby(['cust_class','sex_type'])
multi_group.count()
```

```python
multi_group.mean(numeric_only=True)[['r3m_avg_bill_amt','r6m_B_avg_arpu_amt']]
```

```python
multi_group.max(numeric_only=True)[['r3m_avg_bill_amt']]
```

## 6. 병합

### 1. concat
- default값은 axis = 0

```python
df1 = pd.DataFrame({'key1' : [0,1,2,3,4], 'value1' : ['a', 'b', 'c','d','e']}, index=[0,1,2,3,4])
df2 = pd.DataFrame({'key1' : [3,4,5,6,7], 'value1' : ['c','d','e','f','g']}, index=[3,4,5,6,7])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2]) #axis=0
```

```python
pd.concat([df1,df2],axis=1)
```

```python
df_concat = pd.concat([df1,df2])
df_concat
```

```python
df_concat.reset_index() #기존 인덱스는 새로 col만들고 새로 순차적 인덱스 생성
```

```python
df_concat.reset_index(drop=True) #기존 인덱스는 삭제하고 새로 순차적 인덱스 생성
```

### 2. merge

```python
customer = pd.DataFrame({'cust_id' : [0,1,2,3,4,5],
                    'name' : ['철수', '영희', '길동', '영수', '수민', '동건'],
                    '나이' : [40, 20, 21, 30, 31, 18]})

orders = pd.DataFrame({'cust_id' : [1, 1, 2, 2, 2, 3, 3, 1, 4, 9],
                    'item' : ['치약', '칫솔', '이어폰', '헤드셋', '수건', '생수', '수건', '치약', '생수', '케이스'],
                    'quantity' : [1, 2, 1, 1, 3, 2, 2, 3, 2, 1]})
```

```python
customer
```

```python
orders
```

```python
pd.merge(customer, orders, on='cust_id')
```

```python
pd.merge(customer, orders, on='cust_id', how='inner')
```

```python
pd.merge(customer, orders, on='cust_id', how='left')
```

## <b>연습문제2) 영희가 가장 많이 구매한 아이템은?</b>
   - Hint
     - 1.우선 사람과 아이템 별로 sum을 해서(groupby시 "이름"과 "아이템"기준으로 합을 구하고)
     - 2.loc을 활용하여 (영희의 row의 quantity만 확인)

```python
new = pd.merge(customer,orders,on='cust_id',how='left')
new
```

```python
new1 = new.groupby(['name','item'])['quantity'].sum().reset_index()
new1 = new1.astype({'quantity': int})
new1
```

```python
younghee = new1[new1['name']=='영희'].sort_values('quantity',ascending=False).reset_index(drop=True)
younghee
```

```python
print(younghee.loc[0,'quantity'])
```

```python
print(customer)
print(orders)
```

```python
new_data = pd.merge(customer,orders,on='cust_id')
```

```python
new_data
```

```python
item_summary = new_data.groupby(['name', 'item'])['quantity'].sum().reset_index()
```

```python
item_summary
```

```python
Young_buy = item_summary.loc[item_summary['name']=='영희']
Young_buy
```

```python
Young_buy_sort = Young_buy.sort_values('quantity',ascending=False).reset_index(drop=True)
Young_buy_sort
```

```python
Young_buy_sort.iloc[0],['quantity']
```

```python
print(Young_buy_sort.loc[0, 'quantity'])
```
