### 수치형 데이터

`describe`함수를 통해 기술 통계 확인

```python
cust.describe()
```

이상치 제거하는 함수
q1과 q3의 차 * 1.5 = **Outlier**

```python
def removeOutliers(x, column):
    q1 = x[column].quantile(0.25)
    q3 = x[column].quantile(0.75)
    
    iqr = 1.5 * (q3 - q1)
    
    y = x[(x[column] < (q3 + iqr)) & (x[column] > (q1 - iqr))]
    return(y)
```

```python
cust_data = removeOutliers(cust,'avg_bill')
cust_data=removeOutliers(cust_data, 'A_bill')
cust_data=removeOutliers(cust_data, 'B_bill')
cust_data.describe()
```

정규 분포를 평균이 0, 분산이 1인 표준 정규분포로 변환

#### Normalization

값들을 0과 1사이의 값으로 변환
- `(x - xmin) / (xmax - xmin)`

```python
normalization_df = (cust_data_num - cust_data_num.min()) / (cust_data_num.max() - cust_data_num.min())
normalization_df.head()
```



## [실습2] 수업용 데이터 이상치 처리와 Feature Engineering 하기
cust 데이터에서 
- 이상치는 제거
- 나이는 5단위로 범주화
- 평균 요금은 5개 구간으로 나누어 새로운 변수를 만들어 데이터를 저장

* 이상치 : InterQuartile Range의 1.5배가 넘는 수 (처리 대상 Feature : avg_bill, A_bill, B_bill)
* by_age : 나이는 0 ~ 4세는 0/ 5 ~ 9세는 5 / 10 ~ 14세는 10으로 5단위 범주화
* bill_rating : 전체 평균 요금을 균등비율로 low / lowmid / mid / midhigh / high 5단계로 구분

```python
def removeOut(x,column):
    q1 = x[column].quantile(0.25)
    q3 = x[column].quantile(0.75)
    
    lqr = 1.5 * (q3 - q1)

    y = x[(x[column] < (q3 + lqr)) & (x[column] > (q1 - lqr))]
    return y
```

```python
cust_data = cust.copy()
```

```python
cust_data.info()
cust_data['avg_bill'].describe()
```

```python
cust_data = removeOut(cust_data,'avg_bill')
cust_data = removeOut(cust_data,'A_bill')
cust_data = removeOut(cust_data,'B_bill')
```

```python
cust_data['avg_bill'].describe()
cust_data['A_bill'].describe()
cust_data['B_bill'].describe()
```

```python
cust_data = cust_data.dropna(subset=['age'])
cust_data
```

```python
cust_data['by_age'] = cust_data['age'] // 5 * 5
cust_data = cust_data.astype({'age':int,'by_age':int})
cust_data[['by_age','age']].head()
```

```python
cust_data['bill_rating'] = pd.qcut(cust_data["avg_bill"], 5 ,
                                   labels=['low','lowmid', 'mid','midhigh','high'])
cust_data[['bill_rating','avg_bill']]
```

```python
bill_rating_dummies = pd.get_dummies(cust_data['bill_rating'],dtype = int)
cust_data = pd.concat([cust_data,bill_rating_dummies],axis=1)
cust_data
```
